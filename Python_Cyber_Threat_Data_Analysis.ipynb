## Generate Healthcare Cybersecurity Threat Data

### Subtask:
Generate a Python list of dictionaries representing healthcare cybersecurity incidents. The data will be randomized but follow realistic distributions for attack types, severity, source/target countries, and other specified fields. Ensure more than 300 records are created and all columns (incident_id, timestamp, attack_type, attack_vector, severity, source_country, target_country, source_ip, device_affected, operating_system, target_system, protocol, port, status, detected_by, impact_level, mitigation_action) are populated.


**Reasoning**:
The user wants to generate a Python list of dictionaries with specific keys and populate it with sample cyber data, following the detailed instructions provided.



```
import random
from datetime import datetime, timedelta

# Define lists for categorical fields
attack_types = [
    "Phishing", "Malware", "DDoS", "SQL Injection", "Ransomware",
    "Insider Threat", "Zero-day Exploit", "Credential Stuffing", "Man-in-the-Middle"
]
attack_vectors = [
    "Email", "Network Vulnerability", "Web Application", "USB Drive",
    "Cloud Misconfiguration", "API Exploitation", "Supply Chain"
]
severities = ["Low", "Medium", "High", "Critical"]
countries = [
    "USA", "China", "Russia", "North Korea", "Iran", "Germany", "UK",
    "India", "Brazil", "Canada", "Australia", "Japan"
]
device_types = [
    "Workstation", "Server", "Medical Device", "Mobile Device",
    "IoT Device", "Network Router"
]
operating_systems = [
    "Windows Server", "Windows 10", "Linux", "macOS", "Android", "iOS"
]
target_systems = [
    "EHR System", "Billing System", "Patient Portal", "Research Database",
    "Network Infrastructure", "Medical Imaging System"
]
protocols = ["HTTP", "HTTPS", "FTP", "SSH", "RDP", "SMTP", "DNS"]
ports = [80, 443, 21, 22, 3389, 25, 53, 8080, 445]
statuses = ["Detected", "Mitigated", "Investigating", "Resolved"]
detected_by_methods = [
    "IDS/IPS", "Antivirus", "SIEM", "User Report", "Threat Intelligence",
    "Manual Audit"
]
impact_levels = [
    "Data Breach", "System Downtime", "Financial Loss", "Reputational Damage",
    "Operational Disruption", "Patient Safety Compromise"
]
mitigation_actions = [
    "Quarantine", "Patching", "Firewall Rule", "Account Lockout",
    "Incident Response Plan", "System Restore", "User Training"
]

# Initialize empty list to store generated records
cyber_incidents_data = []

# Generate at least 350 records
for i in range(1, 351):
    # Generate unique incident_id
    incident_id = f"INC{i:05d}"

    # Generate random timestamp within the last year
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365)
    time_between_dates = end_date - start_date
    random_days = random.randrange(time_between_dates.days)
    random_seconds = random.randrange(60 * 60 * 24)
    timestamp = start_date + timedelta(days=random_days, seconds=random_seconds)

    # Randomly select values for categorical fields
    attack_type = random.choice(attack_types)
    attack_vector = random.choice(attack_vectors)
    severity = random.choice(severities)
    source_country = random.choice(countries)
    target_country = random.choice(countries)
    device_affected = random.choice(device_types)
    operating_system = random.choice(operating_systems)
    target_system = random.choice(target_systems)
    protocol = random.choice(protocols)
    port = random.choice(ports)
    status = random.choice(statuses)
    detected_by = random.choice(detected_by_methods)
    impact_level = random.choice(impact_levels)
    mitigation_action = random.choice(mitigation_actions)

    # Generate a random source_ip address (IPv4 format)
    source_ip = f"{random.randint(1, 254)}.{random.randint(1, 254)}.{random.randint(1, 254)}.{random.randint(1, 254)}"

    # Create a dictionary for the current incident
    incident_record = {
        "incident_id": incident_id,
        "timestamp": timestamp.strftime("%Y-%m-%d %H:%M:%S"),
        "attack_type": attack_type,
        "attack_vector": attack_vector,
        "severity": severity,
        "source_country": source_country,
        "target_country": target_country,
        "source_ip": source_ip,
        "device_affected": device_affected,
        "operating_system": operating_system,
        "target_system": target_system,
        "protocol": protocol,
        "port": port,
        "status": status,
        "detected_by": detected_by,
        "impact_level": impact_level,
        "mitigation_action": mitigation_action
    }

    # Append this incident dictionary to the cyber_incidents_data list
    cyber_incidents_data.append(incident_record)

# Print the first few records to verify structure and content
print("Generated Cyber Incident Data (first 5 records):")
for i in range(5):
    print(cyber_incidents_data[i])

print(f"Total records generated: {len(cyber_incidents_data)}")
```

## Analyze Effectiveness of Mitigation Actions by Attack Status

### Reasoning:
To evaluate the effectiveness of various mitigation actions based on the final `status` of a cyber incident, a cross-tabulation will be performed between `status` and `mitigation_action`. The results will then be visualized using a stacked bar chart. This will illustrate which mitigation actions are more frequently associated with 'Resolved' or 'Mitigated' statuses, as opposed to 'Investigating' or 'Detected' statuses, thereby providing insights into their efficacy.

```
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the count of each mitigation action for each status
mitigation_by_status = df_cyber_incidents.groupby(['status', 'mitigation_action']).size().unstack(fill_value=0)

# Reorder columns for better visualization if needed (e.g., if specific statuses are considered more 'effective')
# For example, to put 'Resolved' and 'Mitigated' first:
# status_order = ['Resolved', 'Mitigated', 'Detected', 'Investigating']
# mitigation_by_status = mitigation_by_status.reindex(status_order, axis=0)

# Plotting the stacked bar chart
plt.figure(figsize=(16, 9))
mitigation_by_status.plot(kind='bar', stacked=True, colormap='Paired', figsize=(16, 9))

plt.title('Distribution of Mitigation Actions by Cyber Attack Status')
plt.xlabel('Attack Status')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Mitigation Action', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
